```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
```

```python
def dft(s):
    n = s.size
    f = []
    for i in range(n):
        cf = 0.
        for j in range(n):
            cf += s[j]*np.exp(-1j*2*np.pi*i*j/n)
        f.append(cf / n)
    return np.array(f)

def inv_dft(s):
    n = s.size
    f = []
    for i in range(n):
        cf = 0.
        for j in range(n):
            cf += s[j]*np.exp(1j*2*np.pi*i*j/n)
        f.append(cf / n)
    return np.array(f)
```

```python
x = np.linspace(0, 6*np.pi, 100)
#w = np.sin(x/2)*(np.cos(x)+.75)
w = np.sin(x)+np.random.normal(0., .3, x.size)
fw = dft(w)
xr = inv_dft(fw)
```

```python
plt.polar(x, w)
plt.show()
```

```python
fig, axs = plt.subplots(3, 1)
axs[0].plot(w)
axs[0].set_title('signal')
axs[1].plot(np.abs(fw))
axs[1].set_title('Fourie')
axs[2].plot(xr.real)
axs[2].set_title('xrec')
plt.show()
```
