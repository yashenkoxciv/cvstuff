```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
from skimage.io import imread
from scipy import fftpack
```

```python
img = imread('imgs/6.jpg', True)
```

```python
plt.imshow(img, cmap='gray')
#plt.axis('off')
plt.title('Source image')
plt.show()
```

```python
ft_img = fftpack.fft2(img)
```

```python
plt.imshow(np.abs(ft_img), norm=LogNorm(vmin=5))
plt.colorbar()
plt.title('Image Fourie transform')
plt.show()
```

```python
keep_fraction = 0.3
ft_tr_img = ft_img.copy()
r, c = ft_tr_img.shape
ft_tr_img[int(r*keep_fraction):int(r*(1-keep_fraction))] = 0
ft_tr_img[:, int(c*keep_fraction):int(c*(1-keep_fraction))] = 0
```

```python
plt.imshow(np.abs(ft_tr_img), norm=LogNorm(vmin=5))
plt.colorbar()
plt.title('Image Fourie transform (truncated)')
plt.show()
```

```python
r_img = fftpack.ifft2(ft_tr_img).real
```

```python
fig, axs = plt.subplots(1, 2)
axs[0].imshow(img, cmap='gray')
axs[0].set_title('Source image')
axs[1].imshow(r_img, cmap='gray')
axs[1].set_title('Reconstructed image')
#plt.figure(figsize=(10, 10))
plt.show()
```
