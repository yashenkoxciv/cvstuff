```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
from skimage.io import imread
from scipy import fftpack
```

```python
img = imread('imgs/7.jpg', True)
```

```python
plt.imshow(img, cmap='gray')
#plt.axis('off')
plt.title('Source image')
plt.show()
```

```python
ft_img = fftpack.fft2(img)
```

```python
plt.imshow(np.abs(ft_img), norm=LogNorm(vmin=5))
plt.colorbar()
plt.title('Image Fourie transform')
plt.show()
```

```python
shift_ft_img = fftpack.fftshift(ft_img)
plt.imshow(np.abs(shift_ft_img), norm=LogNorm(vmin=5))
plt.colorbar()
plt.title('Image Fourie transform')
plt.show()
```

```python
r_img = np.abs(fftpack.ifft2(ft_img))
```

```python
fig, axs = plt.subplots(1, 2)
axs[0].imshow(img, cmap='gray')
axs[0].set_title('Source image')
axs[1].imshow(r_img, cmap='gray')
axs[1].set_title('Reconstructed image')
#plt.figure(figsize=(10, 10))
plt.show()
```
