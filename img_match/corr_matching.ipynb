```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.io import imread
from skimage.filters import gaussian
from scipy.signal import convolve2d
```

```python
img1 = gaussian(imread('imgs/bw1.jpg'), 2.5)
img2 = gaussian(imread('imgs/bw2.jpg'), 2.5)
```

```python
fig, axs = plt.subplots(1, 2)
axs[0].imshow(img1, cmap='gray')
axs[0].axis('off')
axs[1].imshow(img2, cmap='gray')
axs[1].axis('off')
plt.show()
```

```python
print('Path from img2')
#path = img2[80:150, 130:180]
path = img2[30:350, 50:270]

plt.imshow(path, cmap='gray')
plt.axis('off')
plt.show()
```

```python
smgv_img1 = img1 - path.mean()
#smgv_path = img2 - path.mean()
smgv_path = path - path.mean()

plt.imshow(smgv_img1, cmap='gray')
plt.axis('off')
plt.show()

plt.imshow(smgv_path, cmap='gray')
plt.axis('off')
plt.show()
```

```python
c_img1 = convolve2d(smgv_img1, smgv_path, boundary='symm', mode='same')
```

```python
plt.imshow(c_img1, cmap='gray')
plt.axis('off')
plt.show()
```

```python
plt.plot(c_img1.flat)
plt.show()
```

```python
shifted_center = np.argwhere(c_img1.max() == c_img1)[0]
```

```python
plt.imshow(img1, cmap='gray')
plt.imshow(img2, cmap='gray', alpha=.75)
plt.plot(img2.shape[0] // 2, img2.shape[1] // 2, 'ro', markersize=8)
plt.plot(shifted_center[0], shifted_center[1], 'go', markersize=8)
plt.show()
```
