```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.filters import gaussian
from skimage.io import imread
```

```python
img1 = gaussian(imread('imgs/bw1.jpg'))
img2 = gaussian(imread('imgs/bw2.jpg'))
```

```python
fig, axs = plt.subplots(1, 2)
axs[0].imshow(img1, cmap='gray')
axs[0].axis('off')
axs[1].imshow(img2, cmap='gray')
axs[1].axis('off')
plt.show()
```

```python
msev = []
for j in range(1, img1.shape[1]):
    msev.append(np.mean(np.square(img1[:, 0:j] - img2[:, img2.shape[1]-j:img2.shape[1]])))
```

```python
print('MSE min value', np.min(msev), 'at', np.argmin(msev))
plt.plot(msev)
plt.title('MSE value')
plt.show()
```

```python
shift_v = np.argmin(msev)
img1_p = img1[:, 0:shift_v]
img2_p = img2[:, img2.shape[1]-shift_v:img2.shape[1]]
imgs_d = np.square(img1_p - img2_p)
#img1_p.shape, img2_p.shape
plt.imshow(img1_p, cmap='gray')
plt.imshow(img2_p, cmap='gray', alpha=0.75)
plt.imshow(imgs_d, alpha=.35)
plt.show()
```

```python
corv = []
for j in range(1, img1.shape[1]):
    corv.append(np.mean(img1[:, 0:j]*img2[:, img2.shape[1]-j:img2.shape[1]]))
```

```python
print('Correlation max value', np.min(corv), 'at', np.argmax(corv))
plt.plot(corv)
plt.title('Correlation value')
plt.show()
```

```python
plt.imshow(np.square(img2 - img1), cmap='gray')
plt.axis('off')
plt.show()
```
