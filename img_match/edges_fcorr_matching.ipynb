```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.io import imread
from skimage.filters import gaussian
from scipy.signal import convolve2d
from scipy.fftpack import fft2, ifft2, fftshift
from skimage.feature import canny
```

```python
# amplitude-phase correlation
img1 = gaussian(imread('imgs/bw1.jpg')) # gaussian()
img2 = gaussian(imread('imgs/bw2.jpg'))
```

```python
fig, axs = plt.subplots(1, 2)
axs[0].imshow(img1, cmap='gray')
axs[0].axis('off')
axs[1].imshow(img2, cmap='gray')
axs[1].axis('off')
plt.show()
```

```python
img1 = canny(img1, sigma=1)
img2 = canny(img2, sigma=1)
```

```python
fig, axs = plt.subplots(1, 2)
axs[0].imshow(img1, cmap='gray')
axs[0].axis('off')
axs[1].imshow(img2, cmap='gray')
axs[1].axis('off')
plt.show()
```

```python
f_img1 = fft2(img1)
f_img2 = fft2(img2)
```

```python
print('Fourie image')

fig, axs = plt.subplots(1, 2)
axs[0].imshow(
    np.log10(np.abs(f_img1)),
    cmap='gray'
)
axs[0].set_title('img1 Fourie image')
axs[0].axis('off')
axs[1].imshow(
    np.log10(np.abs(f_img2)),
    cmap='gray'
)
axs[1].set_title('img2 Fourie image')
axs[1].axis('off')
plt.show()

fig, axs = plt.subplots(1, 2)
axs[0].imshow(
    fftshift(np.log10(np.abs(f_img1))),
    cmap='gray'
)
axs[0].set_title('img1 Fourie image shifted')
axs[0].axis('off')
axs[1].imshow(
    fftshift(np.log10(np.abs(f_img2))),
    cmap='gray'
)
axs[1].set_title('img1 Fourie image shifted')
axs[1].axis('off')
plt.show()
```

```python
corrf = f_img1*np.conjugate(f_img2)
```

```python
plt.imshow(
    fftshift(np.log10(np.abs(corrf))),
    cmap='gray'
)
plt.title('Cross-correlation')
plt.axis('off')
plt.show()
```

```python
rcorrf = ifft2(corrf)
```

```python
ccfr = fftshift(np.abs(rcorrf))
```

```python
plt.imshow(
    ccfr,
    cmap='gray'
)
plt.title('Cross-correlation field reconstruction shifted')
plt.axis('off')
plt.show()

plt.imshow(
    np.abs(rcorrf),
    cmap='gray'
)
plt.title('Cross-correlation field reconstruction')
plt.axis('off')
plt.show()
```

```python
shifted_center = np.argwhere(ccfr.max() == ccfr)[0]
shifted_center
```

```python
plt.imshow(img1, cmap='gray')
plt.imshow(img2, cmap='gray', alpha=.75)
plt.plot(img1.shape[0] // 2, img1.shape[1] // 2, 'ro', markersize=8)
plt.plot(shifted_center[1], shifted_center[0], 'go', markersize=8)
plt.show()
```
