```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.io import imread
from skimage.filters import gaussian
```

```python
img1 = gaussian(imread('imgs/bw1.jpg'), 1.5)
img2 = gaussian(imread('imgs/bw2.jpg'), 1.5)
```

```python
fig, axs = plt.subplots(1, 2)
axs[0].imshow(img1, cmap='gray')
axs[0].axis('off')
axs[1].imshow(img2, cmap='gray')
axs[1].axis('off')
plt.show()
```

```python
def image_entropy(img):
    ''' Instead of np.histogram np.unique can be used:
            v, c = np.unique(img, return_counts=True)
        For testing use skimage.measure.shannon_entropy
    '''
    c, _ = np.histogram(img, 100)
    pk = c / np.sum(c)
    return -np.sum(pk*np.log2(pk))

image_entropy(img1)
```
